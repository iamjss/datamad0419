### ¡Análisis y Limpieza de datos!

A simple vista podemos comprobar que hay columnas con muchos datos nulos, por lo que vamos a ver si efectivamente hay columnas con un número significativo de Nulls y en ese caso las eliminamos de nuestro análisis.

```python
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

Analizando las variables con un mayor número de datos nulos, considero que la variable edad y time (una vez limpiadas) pueden ser significativas, por lo que partiré eliminando las variables Unnamed que están vacías.

```python
drop_cols = list(null_cols[null_cols > 3500].index)
data = data.drop(drop_cols, axis=1)
```

```python
data.head(10)
```

Con **.info** obtenemos la información del dataset, el número de columnas, el número de datos dentro de cada columna y qué tipo de datos tiene.

```python
data.info()
```

Elimino las columnas que sé que no me van a interesar en absoluto, ya sea porque dan los mismos valores o porque son valores que no influyen en el dataset y reducimos nuestro dataset de 22 a 11 columnas.

```python
before = len(data.columns)
select_columns = ['Case Number', 'Date', 'Area', 'Location', 'Name', 'Species ', 'Investigator or Source', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2', 'original order']

data = data.drop(select_columns, axis=1)
after = len(data.columns)
print('Número de columnas finales: ', after)
```

```python
data.head(10)
```

Arreglamos los datos de la variable 'Age'.

```python
data['Age'] = data['Age'].fillna('0')
```

Mostramos los nulos de la variable 'Sex'.

```python
null_displ = data[(data['Sex '].isnull()==True)]
null_displ = null_displ[['Year', 'Type', 'Country',
       'Activity', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)', 'Time']]
null_displ
```

Mostramos los elementos que conforman la variable 'Sex' y cambiamos de nombre la columna para eliminarle el espacio que tenía.

```python
data[('Sex ')].value_counts()
```

```python
data = data.rename(columns={'Sex ':'Sex'})
data.columns
```

Dado que no nos interesa ni transformar los nulos ni los valores que no son Mujer, filtramos los datos que tenemos acordes a la Variable Sexo-Mujer(F).

```python
data = data[(data['Sex']=='F')]

data
```

Hacemos filtering de los datos. Primero en la variable 'Age' creamos una función para eliminar todo lo que no sea dígito de los valores para poder cambiarle la clase/type a la variable de *object* a *int64*.

```python
print(set(data['Age']))
```

```python
def no_letras(valor):
    pattern = '[^0-9]*'
    number = re.sub(pattern,"",valor)
    if len(number) > 2: number = number[:2]
    return number
data['Age'] = data['Age'].apply(no_letras)
print(set(data['Age']))
```

Como hemos creado valores vacíos, los equiparamos a 0.

```python
data.loc[(data['Age']==""), 'Age']=0
data['Age'] = data['Age'].astype('int64')
data.head(10)
```

Hallamos la media de los valores en 'Age' y la asignamos a cada valor que sea 0.

```python
meanAge = int(round(data['Age'][(data['Age']>0)].mean()))
meanAge
```

```python
data.loc[(data['Age']==0), 'Age']=meanAge
data.head(10)
```

Analizando la variable 'Year', comprobamos si hay nulos y si es así eliminamos las filas en las que dan valor nulo.

```python
data[('Year')].value_counts()
```

```python
data = data[data.Year != 0]
```

Analizando la variable 'Type' de tipos de ataque, comprobamos si hay nulos y si es así eliminamos las filas en las que dan valor nulo.

```python
data[('Type')].value_counts()
```

En este caso lo que tenemos son valores 'Invalid' que delimitan que falta información y se podrían considerar nulos por lo que eliminamos las filas en las que ocurren estos ataques.

```python
data = data[data.Type != 'Invalid']
data.head(10)
```

### Extraemos el nuevo Dataset en un formato .csv
