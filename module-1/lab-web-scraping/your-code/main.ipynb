####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
print(soup.prettify())
````

````python
table = soup.find_all('tr', 'normal')
print(table)
````

````python
earthquake = []

for quake in table:
    x = quake.contents[3].find('a')
    date = quake.contents[3].find('a').string.split()[0]
    time = quake.contents[3].find('a').string.split()[1]
    latitude = quake.contents[4].string + quake.contents[5].string
    longitude = quake.contents[6].string + quake.contents[7].string
    region = quake.contents[11].string
    earthquake.append([date, time, latitude, longitude, region])

df = pd.DataFrame(earthquake, columns = ['Date', 'Time', 'Latitude', 'Longitude','Region'])

df.head(10)
````
