```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize

client = MongoClient ('localhost', 27017)
db = client['companies']
```

Mi objetivo es ver si es posible la ubicación de nuestra empresa de videojuegos buscando el mercado europeo y localizada en algún lugar de Europa.

```python
a = db.companies.find({'$and':[{'category_code': 'games_video'},{'offices.latitude': {'$ne': 'null'}},{'offices.longitude': {'$ne': 'null'}},{'$or':[{'offices.country_code':'ESP'},{'offices.country_code':'IRL'},{'offices.country_code':'FRA'},{'offices.country_code':'GBR'},{'offices.country_code':'GER'},{'offices.country_code':'NLD'},{'offices.country_code':'ITA'},{'offices.country_code':'PRT'},{'offices.country_code':'CHE'},{'offices.country_code':'AUT'},{'offices.country_code':'NOR'},{'offices.country_code':'SWE'}]}]})
for _ in a:
    print(_['name'])
```

Convierto el resultado de la query en un Dataframe para poder trabajar con él.

```python
eurdf = pd.DataFrame(db.companies.find({'$and':[{'category_code': 'games_video'},{'offices.latitude': {'$ne': 'null'}},{'offices.longitude': {'$ne': 'null'}},{'$or':[{'offices.country_code':'ESP'},{'offices.country_code':'IRL'},{'offices.country_code':'FRA'},{'offices.country_code':'GBR'},{'offices.country_code':'GER'},{'offices.country_code':'NLD'},{'offices.country_code':'ITA'},{'offices.country_code':'PRT'},{'offices.country_code':'CHE'},{'offices.country_code':'AUT'},{'offices.country_code':'NOR'},{'offices.country_code':'SWE'}]}]})
)
eurdf.head()
```

Hago un flatten de la columna offices para poder obtener las oficinas fuera de una lista y así a su vez poder, posteriormente, sacar del diccionario correspondiente la ubicación de las empresas en la ciudad de Londres.

```python
def flattencolumns(df1, cols):
    eurdf = pd.concat([pd.DataFrame(df1[x].values.tolist()).add_prefix(x) for x in cols], axis=1)
    return pd.concat([eurdf, df1.drop(cols, axis=1)], axis=1)
```

```python
cols = ['offices']
df = flattencolumns(eurdf, cols)
df.head(10)
```
