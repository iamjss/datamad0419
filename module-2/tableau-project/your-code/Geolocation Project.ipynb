```python
from pymongo import MongoClient
import pandas as pd
from pandas.io.json import json_normalize

client = MongoClient ('localhost', 27017)
db = client['companies']
```

Mi objetivo es ver si es posible la ubicación de nuestra empresa de videojuegos buscando el mercado europeo y localizada en algún lugar de Europa.

Viendo que hay tantas empresas en Reino Unido, vamos a comprobar cuántas puede haber en la ciudad de Londres.

```python
for _ in db.companies.find({'$and':[{'category_code': 'games_video'},{'offices.city':'London'},{'offices.latitude': {'$ne': 'null'}},{'offices.longitude': {'$ne': 'null'}}]}):
    print("Name of the company: {}.".format(_["name"]))
```

Hago un flatten de la columna offices para poder obtener las oficinas fuera de una lista y así a su vez poder, posteriormente, sacar del diccionario correspondiente la ubicación de las empresas en la ciudad de Londres.

```python
def flattencolumns(df1, cols):
    medialon = pd.concat([pd.DataFrame(df1[x].values.tolist()).add_prefix(x) for x in cols], axis=1)
    return pd.concat([medialon, df1.drop(cols, axis=1)], axis=1)
```
