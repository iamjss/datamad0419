```python
import pandas as pd
```

```python
def dataset():
    df = pd.read_csv('tableau_project.csv')
    return df
df= dataset()
df.head(10)
```

```python
df.columns
```

```python
df = df.drop(columns=['journey_id'])
df.head()
```

```python
df = df.rename(index=str, columns={'Unnamed: 0': 'id', 'vehicle_type_id': 'vehicle_id', 'start_type': 'type', 
                              'start_at': 'start_time', 'start_lat': 'latitude1', 'start_lon': 'longitude1', 'end_at': 'end_time', 
                             'end_lat': 'latitude2', 'end_lon': 'longitude2', 'end_state': 'end'})
df.head()
```

```python
df.info()
```

```python
null_cols = df.isnull().sum()
null_cols[null_cols > 0]
```

```python
df[('discount')].value_counts()
```

```python
df[('price_supplements')].value_counts()
```

```python
def dropping(df):
    drop_cols = list(null_cols[null_cols > 1000].index)
    df = df.drop(drop_cols, axis=1)
    drop_cols2 = ['price_supplements', 'discount']
    df = df.drop(drop_cols2, axis=1)
    return df
df = dropping(df)
df.head(10)
```

```python
drop_cols2 = ['price_supplements', 'discount']
df = df.drop(drop_cols2, axis=1)
df.head()
```

```python
def nulos(df):
    df['latitude2'] = df['latitude2'].fillna('0')
    df['longitude2'] = df['longitude2'].fillna('0')
    df = df[df.latitude2 != '0']
    df = df[df.longitude2 != '0']
    return df
df = nulos(df)
df.head()
```

```python
df.to_csv('tableau_project_cleaned.csv')
```
